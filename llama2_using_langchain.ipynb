```python
from langchain.llms import HuggingFacePipeline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# **Loading the Model**

```python
model = "meta-llama/Llama-2-7b-hf"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model = model,
    tokenizer = tokenizer,
    torch_dtype = torch.bfloat16,
    trust_remote_code = True,
    device_map = "auto",
    max_length = 1000,
    do_sample = True,
    top_k = 10,
    num_return_sequences = 1,
    eos_token_id = tokenizer.eos_token_id
)
```

```python
pipeline
```

```python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature': 0.5})
```

```python
prompt = "What might be a good name for a company that sells umbrellas."
```

```python
print(llm(prompt))
```

```python
prompt2 = "I am thinking of starting a new restaurant chain, can you give me some tips?"
```

```python
print(llm(prompt2))
```

# **Creating a custom prompt template**

```python
from langchain.prompts import PromptTemplate
```

```python
from langchain.chains import LLMChain
```

```python
prompt_template1 = PromptTemplate(
    input_variables = ['cuisine'],
    template = "I want to open a restaurant chain to sell {cuisine} food. Suggest me a few extravagant names. Also give me some tips for the same."
)
```

```python
input_prompt = prompt_template1.format(cuisine = "Italian")
```

```python
print(input_prompt)
```

```python
chain = LLMChain(llm = llm, prompt = prompt_template1, verbose = True)
```

```python
response = chain.run("Indian")
```

```python
print(response)
```

```python
response2 = chain.run("Italian")
```

```python
print(response2)
```
